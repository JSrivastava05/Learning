- np.char.rpartition('hello world numpy rpartition',sep=' ')

- np.char.less(s1,s2) # s1<s2

- np.char.count(a,substr,start,end=None)
- np.char.startswith(str/arr,substr/substrarray,start=0,end=None)
- np.char.endswith(str/arr,substr/substrarray,start=0,end=None)
- np.char.find(str/arr,sub/substrarr,start,end)
- np.char.rfind('hello world','lo')
- np.char.index(['hello world','findmefindme'],['wo','findme'],start=0)) # first occurance
- np.char.rindex(['hello world','findmefindme'],['l','findme'],start=0)) # [9 6]
- np.char.isalpha([1,'1','asd98765','jhgfd']) # True iff all chars are alpha 
- np.char.isalnum([1,'1','asd98765','jhgfd']) # True if alpha or num or alphanumeric
- np.char.isdecimal([1,1.2,'18765','D','1.2','2/3','asd98765','jhgfd','56kjhg'])) # True if num or num as string
- np.char.isdigit([1,1.2,'18765','D','1.2','2/3','asd98765','jhgfd','56kjhg']) # same as isdecimal?
- np.char.isnumeric([1,1.2,'18765','D','1.2','2/3','asd98765','jhgfd','56kjhg'])) #return True if there are only numeric characters in the element.
- np.char.islower(['a','Aa','1','1.2']) #[ True False False False]
- np.char.isupper(['A','ASD'])
- np.char.isspace(['a a','  a','  ',' ajksdfh']) #[False False  True False]

- np.char.istitle('I Am Learning NUMPY') #f
- np.char.istitle('I Am Learning Numpy') #t
- np.char.str_len(str)
- np.char.array(['1','a','ajkdhfjk'],itemsize=5, copy=True, unicode=None, order=None) # itemsize keeps only 5 chars of last item
- np.chararray((3,3))) # blank array of ''
- np.chararray(shape, itemsize=1, unicode=False, buffer=None, offset=0, strides=None, order=None)


```python
import numpy as np
a,b='Hello', 'world'
print('np.char.add() : ', np.char.add(a,b)) # Concatenate is for integer, will not work for string
a,b=['John','Jane'], ['Doe','Dawson']
print(np.char.add(a,b))
print(np.char.multiply('Numpy ',3))
print(np.char.center('Numpy',20,fillchar='*'))
print(np.char.capitalize(['numpy python','numpy','python'])) 
print(np.char.lower(['NUMPY Python','NUMpy','PYthon']))
print(np.char.upper(['numpy python','numpy']))
print(np.char.title(['i am studying numpy python']))
print(np.char.split('i am studying numpy python', sep=' '))
print(np.char.split('i,am,studying,numpy,python', sep=','))
print(np.char.split('i_am_studying_numpy_python', sep='_'))
# print(np.char.lsplit('hello world numpy lsplit',sep=' '))
print(np.char.rsplit('hello world numpy rsplit',sep=' ',maxsplit=2))
print(np.char.partition('hello world',sep=' '))
# print(np.char.lpartition('hello world numpy lpartition',sep=' ')) # fn doesnt exists
print(np.char.rpartition('hello world numpy rpartition',sep=' ')) # partition from right side of str
print(np.char.splitlines('one line \n two line \n three line' ))
print(np.char.splitlines('one line \r two line \r three line' ))
print(np.char.splitlines('one line \r\n two line \r\n three line' ))
print(np.char.strip('numpy python','p')) # nothing happens
print(np.char.strip('numpy python','n')) #characters leading and/or trailing in it
print(np.char.strip('python numpy','py'))
print(np.char.lstrip('python numpy','py'))
print(np.char.rstrip('python numpy','py'))
print(np.char.join(':','dmy'))
print(np.char.join([':',':',"/","-"],['hhmmss','hms','ymd','dmy'])) # adding separator after each char
print(np.char.replace('I am learning numpy python','learning', 'studying'))
print(np.char.replace('I am learning numpy python','a', 'b'))
a='hello'
a=np.char.encode(a,'cp500')
print(a)
print(np.char.decode(a,'cp500' ))
print('hello\tworld')
print(np.char.expandtabs('hello\tworld',tabsize=16))
print(np.char.ljust('hello',width=20,fillchar='*'))
print(np.char.rjust('hello',width=20,fillchar='*'))
print(np.char.swapcase('HeLlo wOrLd'))
# print(np.char.translate('hello world','*'))
print(np.char.zfill('xxx',width=18))
print(np.char.equal('z','y'))
print(np.char.equal(['abc','zyx'],['abc','cba'])) #Return (x1 == x2) element-wise.
print(np.char.not_equal(['abc','xyz'],['abc','xyy']))
print(np.char.greater_equal('a a ','a a     ')) #  this comparison is performed by first stripping whitespace characters from the end of the string.
print(np.char.less_equal('a a  a','a aa'))  
print(np.char.greater('abc','a')) # first string greater than second
print(np.char.greater('abc','abcd'))
print(np.char.less('a','aa')) # second string greater than first
print(np.char.less('aaa','aa'))
print(np.char.compare_chararrays(['a','b'],['a','c'],cmp="==",rstrip=True))
print(np.char.compare_chararrays(['a','b'],['a','c'],cmp="!=",rstrip=True))
print(np.char.compare_chararrays(['ab','bc'],['a','c'],cmp=">=",rstrip=True))
print(np.char.compare_chararrays(['a','bc'],['a','c'],cmp="<=",rstrip=True))
print(np.char.compare_chararrays(['ab','cb'],['a','c'],cmp=">",rstrip=True))
print(np.char.compare_chararrays(['a','b'],['aa','cb'],cmp="<",rstrip=True))
print(np.char.count('hello world','l',0))
print(np.char.count('((()))))(())',[')','(']))
print(np.char.count(['abcbc','bbbbc','ccc'],['a','b','c'],0))
print(np.char.startswith(['hello python','numpy','numpy'],['hell','jsh','nu'],0))
print(np.char.endswith(['hello python','numpy','numpy'],['on','py','ing'],0))
print(np.char.find('hello world','lo',0)) # returns index from 0
# print(np.char.lfind('hello world','lo')) # no such func
print(np.char.rfind('hello world','lo'))# returns index from 0
print(np.char.index('hello world','wo',start=0))
print(np.char.index(['hello world','findmefindme'],['wo','findme'],start=0)) # first occurance
print(np.char.rindex(['hello world','findmefindme'],['l','findme'],start=0)) # [9 6]
print(np.char.isalpha([1,'1','asd98765','jhgfd'])) # True iff all chars are alpha 
print(np.char.isalnum([1,'1','asd98765','jhgfd'])) # True if alpha or num or alphanumeric
print(np.char.isdecimal([1,1.2,'18765','D','1.2','2/3','asd98765','jhgfd','56kjhg'])) # True if num or num as string, return True if there are only decimal characters in the element.
# [ True False  True False False False False False False]
print(np.char.isdigit([1,1.2,'18765','D','1.2','2/3','asd98765','jhgfd','56kjhg'])) # same as isdecimal?
# [ True False  True False False False False False False]
print(np.char.isnumeric([1,1.2,'18765','D','1.2','2/3','asd98765','jhgfd','56kjhg'])) #return True if there are only numeric characters in the element.
# [ True False  True False False False False False False]
print(np.char.islower(['a','Aa','1','1.2'])) #[ True False False False]
print(np.char.isupper(['A','ASD']))
print(np.char.isspace(['a a','  a','  ',' ajksdfh'])) #[False False  True False]
print(np.char.istitle('I am learning Numpy')) #f
print(np.char.istitle('I Am Learning NUMPY')) #f
print(np.char.istitle('I Am Learning Numpy')) #t
print(np.char.str_len(['1','1.2','abcd']))
print(np.char.array('hello world'))
print(np.char.array(['1','a','ajkdhfjk'],itemsize=5, copy=True, unicode='cp500', order=None)) # itemsize keeps only 5 chars of last item
print(np.char.asarray(['14jk','sadjbfjsdjfkj','ajkdhfjk'],itemsize=3, unicode=None, order=None)) # same as above copy if necessary
print(np.chararray((3,3))) # blank array of ''
a=np.chararray((3,3))
a[:]='hi numpy'
print(a) # matrix of only first char due to default itemsize 1
a=np.chararray((3,3),itemsize=5)
a[:]='NumpyPython'
print(a)
```
