```python
# find value in tree
class bt:
    def __init__(self,data=None):
        self.value=data
        self.left=None
        self.right=None
    def insert(self,data):
        if self.value==None:
            self.value=data
        else:
            if data<self.value:
                if self.left==None:
                    self.left=bt(data)
                else:
                    self.left.insert(data)
            if data>self.value:
                if self.right==None:
                    self.right=bt(data)
                else:
                    self.right.insert(data)
    def printtree(self):
        if self.value!=None:
            if self.left!=None:
                self.left.printtree()
            print(self.value)
            if self.right!=None:
                self.right.printtree()
    def findval(self,data):
        if self.value==None:
            return print('tree is empty')
        else:
            if data<self.value:
                if self.left==None:
                    return print(data, ' not found')
                self.left.findval(data)
            elif data> self.value:
                if self.right==None:
                    return print(data, ' not found')
                self.right.findval(data)
            else:
                print(data, ' is found')
                    
if __name__=='__main__':
    root=bt()
    root.findval(18)
    root=bt(10)
    root.insert(7)
    root.insert(13)
    root.insert(12)
    root.insert(5)
    root.findval(5)
    root.insert(15)
    root.insert(3)
    root.insert(6)
    print('tree -')
    root.printtree()
    root.findval( 6)
    root.findval( 18)
```

```python
# min,max value in tree
class bt :
    def __init__(self,data=None):
        self.value=data
        self.left=None
        self.right=None
    def insert(self,data):
        if self.value==None:
            self.value=data
        else:
            if data<self.value:
                if self.left==None:
                    self.left=bt(data)
                else:
                    self.left.insert(data)
            if data>self.value:
                if self.right==None:
                    self.right=bt(data)
                else:
                    self.right.insert(data)
    def printtree(self):
        if self.value==None:
            return None
        else:
            res=[]
            if self.left!=None:
                res=self.left.printtree()
            res.append(self.value)
            if self.right!=None:
                res=res+ self.right.printtree()
            return res
    def inorder(self):
        #LRR
        if self.value==None:
            return None
        else:
            res=[]
            if self.left!=None:
                res=self.left.inorder()
            res.append(self.value)
            if self.right!=None:
                res=res+self.right.inorder()
            return res
    def preorder(self):
        #RLR
        if self.value==None:
            return None
        else:
            res=[]
            res.append(self.value)
            if self.left!=None:
                res=res+self.left.preorder()
            if self.right!=None:
                res=res+self.right.preorder()
            return res
    def postorder(self):
        #LRR
        if self.value==None:
            return None
        else:
            res=[]
            if self.left!=None:
                res=self.left.postorder()
            if self.right!=None:
                res=res+ self.right.postorder()
            res.append(self.value)
            return res
                
    def findval(self,data):
        if self.value==None:
            print('tree is empty ')
        else:
            if data<self.value:
                if self.left==None:
                    print(data , ' isnt found')
                else:
                    self.left.findval(data)
            elif data>self.value:
                if self.right==None:
                    print(data , ' isnt found')
                else: 
                    self.right.findval(data)
            else:
                print(data , ' is found')
            
    def minval(self):
        # min will always be in left subtree
        if self.value==None:
            return None
        else:
            minival=self.value
            if self.left!=None:
                minival=self.left.minval()
            return minival
    def maxval(self):
        if self.value==None:
            return None
        else:
            maxival=self.value
            if self.right!=None:
                maxival=self.right.maxval()
            return maxival
            
if __name__=='__main__':
    root=bt()
    print('First iteration - tree is empty')
    print('Inorder - ', root.inorder())
    print('Preorder - ', root.preorder())
    print('Postorder - ', root.postorder())
    print('find value 7 -', end=' ')
    root.findval(7)
    print('Minimum value - ', root.minval())
    print('Maximum value - ', root.maxval())
    root.insert(10)
    root.insert(7)
    root.insert(13)
    root.insert(12)
    root.insert(5)
    print('\nSecond iteration - tree is filled with 10, 7 ,13, 12, 5')
    print('Inorder - ', root.inorder())
    print('Preorder - ', root.preorder())
    print('Postorder - ', root.postorder())
    print('Minimum value - ', root.minval())
    print('Maximum value - ', root.maxval())
    print('find value 1 -', end=' ')
    root.findval(1)
    root.insert(15)
    root.insert(3)
    root.insert(6)
    print('\nThird iteration - tree is filled with 10, 7 ,13, 12, 5, 15, 3, 6')
    print('tree - ', root.printtree())
    print('Inorder - ', root.inorder())
    print('Preorder - ', root.preorder())
    print('Postorder - ', root.postorder())
    print('find value 7 -', end=' ')
    root.findval(7)
    print('Minimum value - ', root.minval())
    print('Maximum value - ', root.maxval())



                    
```

```python
# average value of tree per level
from queue import Queue
class bt:
    def __init__(self,data=None):
        self.value,self.left,self.right=data,None,None
    def printtree(self):
        if self.value==None:
            return None
        else:
            res=[]
            if self.left!=None:
                res=self.left.printtree()
            res.append(self.value)
            if self.right!=None:
                res=res+self.right.printtree()
            return res

    
    def avg_by_level_BFS(self):
        q=Queue()
        q.put(self) # -> add to queue
        while not q.empty():
            sm=0
            cnt=0
            tmp=Queue() # -> Create new queue instance
            while not q.empty(): #q.empty()-> checks if queue is empty
                n=q.queue[0] # -> fifo, first elem in n
                q.get() # > remove first elem from queue
                sm=sm+n.value
                cnt=cnt+1
                if n.left!=None:
                    tmp.put(n.left) #-> adding left node in tmp
                if n.right!=None:
                    tmp.put(n.right) #-> adding left node in tmp
            q=tmp
            print(sm/cnt,end=' ')
            
    def min_max_by_level_BFS(self):
        q=Queue()
        q.put(self)
        while not q.empty():
            tmp=Queue()
            lst=[]
            while not q.empty():
                n=q.queue[0]
                lst.append(n.value)
                q.get()
                if n.left!=None:
                    tmp.put(n.left)
                if n.right!=None:
                    tmp.put(n.right)
            q=tmp
            print('Min_val - ', min(lst),'MaxVal - ', max(lst))
            
if __name__=='__main__':
    root=bt(10)
    root.left=bt(7)
    root.right=bt(13)
    root.left.left=bt(5)
    root.left.left.left=bt(3)
    root.left.left.right=bt(6)
    root.right.left=bt(12)
    root.right.right=bt(15)
    res=root.printtree()
    print(res)
    root.avg_by_level_BFS()
    root.min_max_by_level_BFS()
    
```
